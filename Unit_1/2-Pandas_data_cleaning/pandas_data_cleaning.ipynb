```python
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.txt', sep = '\t')
```

```python
data = pd.concat([file1,file2])
data
```

```python
data.columns = data.columns.str.lower()
data
```

```python
data = data.rename(columns={'controln':'id',
                            'hv1':'median_home_val',
                            'ic1':'median_household_income'})
```

```python
#When there are outlieres (just a few high numbers) in the column --> the best is to fill NaN with the median of this column
ic4_mean = data['ic4'].mean()
data['ic4'] = data['ic4'].fillna(ic4_mean)
```

```python
data[data['ic4'].isna()==True]
```

```python
# Replacing null values for categorical variables

data['gender'].value_counts()
# len(data[data['gender'].isna()==True])  # number of missing values
```

```python
data.columns
```

#### replace missing values in multiple columns

```python
replace_dict = {
    "gender": "F",
    "state": "NY"
}

for colum in data.columns:
    data[column] = data[column].fillna(replace_dict[colum])
```

```python
data['gender'].value_counts()
```

```python
data['gender'] = data['gender'].fillna('F') #with 'F' because it's the most common one
# len(data[data['gender'].isna()==True]) # now this number is 0
```

```python
# Exporting this processed data to a csv
data.to_csv('merged_clean_ver9.csv')
#if I want a completely new file --> just change the name 
```

```python
data['gender'].unique() # check the unique values in the column
```

```python
# data = data.dropna()
data = data.dropna(subset=['gender'])
```

```python
def clean(x):
    if x.upper().startswith('M'):
        return 'Male'
#     elif x.startswith('M'):
#         return 'Male'
    elif x.lower().startswith('f'):
        return 'Female'
    else:
        return 'Other'
    
data['gender'] = data['gender'].apply(clean)

data['gender'].unique()
```
